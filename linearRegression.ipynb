```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('USA_Housing.csv')
```

```python
dataset.head()
```

```python
X = dataset.drop(['Price', 'Address'], axis=1)
y = dataset['Price']
```

```python
pipeline = Pipeline([
    ('scaler', StandardScaler())
])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_train_scaled = pipeline.fit_transform(X_train)
X_test_scaled = pipeline.transform(X_test)
```

```python
model_lr = LinearRegression()
model_lr.fit(X_train_scaled, y_train)
```

```python
model_rf = RandomForestRegressor(n_estimators=100, random_state=42)
model_rf.fit(X_train_scaled, y_train)
```

```python
y_train_pred_lr = model_lr.predict(X_train_scaled)
y_test_pred_lr = model_lr.predict(X_test_scaled)

train_score_lr = r2_score(y_train, y_train_pred_lr)
test_score_lr = r2_score(y_test, y_test_pred_lr)
train_mse_lr = mean_squared_error(y_train, y_train_pred_lr)
test_mse_lr = mean_squared_error(y_test, y_test_pred_lr)
```

```python
print("Linear Regression - Train R2 Score:", train_score_lr)
print("Linear Regression - Test R2 Score:", test_score_lr)
print("Linear Regression - Train MSE:", train_mse_lr)
print("Linear Regression - Test MSE:", test_mse_lr)
```
